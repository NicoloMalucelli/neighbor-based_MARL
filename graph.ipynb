```python
from matplotlib import pyplot as plt
import numpy as np

nbrs = np.array([i for i in range(1, 11)])
nn_weight = 2.80 # MB 2,80 MB (2.941.868 byte)
trajectory_weight = 2.10 #876 * 300 = 262800 # 300 = episode length

nn_consensus = np.full_like(nbrs, nn_weight)
nn_averaging = nbrs * nn_weight + nbrs * 6.4e-5
experience_sharing = nbrs * trajectory_weight

plt.figure(figsize=(10, 6))

plt.plot(nbrs, nn_consensus, label='NN-Consensus')
plt.plot(nbrs, nn_averaging, label='NN-Averaging')
plt.plot(nbrs, experience_sharing, label='Experience Sharing')

# Add labels and title
plt.xlabel('Number of Neighbors')
plt.ylabel('Weight (MB)')
plt.title('Communication overhead')
plt.legend()
```
